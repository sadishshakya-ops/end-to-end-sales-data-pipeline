```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Admin\Desktop\Capstone Project\Sample - Superstore cleaned.csv")

print(df.info())
print(df.describe())

# Sales by Category
df.groupby("Category")["Sales"].sum().plot(kind="bar", color="lightblue", edgecolor="black")
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()

# Profit Trend
df.groupby("Order_Year")["Profit"].sum().plot(marker="o", linestyle="-", color="blue")
plt.title("Yearly Profit Trend")
plt.xlabel("Year")
plt.ylabel("Total Profit")
plt.show()


```
